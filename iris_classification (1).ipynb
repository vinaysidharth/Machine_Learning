```python
#Loading Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import sklearn as sls

#Import dataset
from sklearn.datasets import load_iris
data=load_iris()
target=data.target
df=pd.DataFrame(data.data,columns=data.feature_names)
df['class']=target
df.head()
sns.pairplot(df, hue='class', palette='viridis', diag_kind='kde')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02, fontsize=16)
plt.show()

#train-test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.data,data.target)
from sklearn.neighbors import KNeighborsClassifier

n_neighbors_scores = {}

# Loop through a range of n_neighbors values
for n in range(1, 100):  # Adjust the range as needed
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    score = knn.score(X_test, y_test)
    n_neighbors_scores[n] = score
plt.plot(list(n_neighbors_scores.keys()), list(n_neighbors_scores.values()))
plt.xlabel('n_neighbors')
plt.ylabel('Test Score')
plt.title('Test Score vs. n_neighbors')
plt.show()
```
